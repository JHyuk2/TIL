```python
import import_ipynb
import myfunc
import MyModels
import pandas as pd
import numpy as np
```

```python
data = myfunc.read_file("data_sample.xlsx")
```

```python
len(data)
```

```python
ch1 = myfunc.remove_url(myfunc.split_all_sentence(data["Text"]))
```

```python
ch2 = myfunc.change(ch1)
```

```python
new_text = []
for test1 in ch2:
    s1= []
    for test2 in test1.split():
        if test2 != " ":
            s1.append(test2)
    new_text.append(" ".join(s1))
```

```python
Label = data["Label"]
del data
data = pd.DataFrame({"Text" : new_text,"Label" : Label})
```

```python
data_after = data.drop_duplicates(["Text"],keep = "last")
print(len(data))
print(len(data_after))
```

```python
data_after.head()
```

```python
Train_X,Test_X,Train_Y,Test_Y = MyModels.data_transformer(data_after)
```
