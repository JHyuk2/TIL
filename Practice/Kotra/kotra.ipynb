```python
result_dict = {}

for key, val in data.items():    
    if len(val) >= 3:
        count = 0
        for k in val.keys():
            if val[k] >= 50:
                count += 1
        if len(val) == count:
            result_dict[key] =  val

print(result_dict)
print(len(result_dict))
```
