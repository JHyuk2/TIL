```python
import os
directory = "./dataset"
files = os.listdir(directory)
pdf_files = [file for file in files if file.endswith('.pdf')]
```

```python
import PyPDF2

def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text

import fitz  # PyMuPDF
def read_pdf_with_pymupdf(file_path):
    doc = fitz.open(file_path)
    text = ''
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)  # 페이지 불러오기
        text += page.get_text()  # 페이지에서 텍스트 추출
    return text

import pdfplumber
def read_pdf_with_pdfplumber(file_path):
    doc = pdfplumber.open(file_path)
    text = ''
    for page in doc.pages:
        text += page.extract_text() + '\n'
    return text
```

```python
path = "./dataset/"
# file_path = path+pdf_files[0]
import pandas as pd
columns = ["file_name", "pypdf", "pymupdf", "pdfplumber"]
df = pd.DataFrame(columns=columns)

for file_name in pdf_files:
    file_path = path+file_name
    pypdf = read_pdf(file_path)
    pymupdf = read_pdf_with_pymupdf(file_path)
    pdf_pdfplumber = read_pdf_with_pdfplumber(file_path)
    df = pd.concat([df, pd.DataFrame([{"file_name":file_name, "pypdf":pypdf, "pymupdf":pymupdf, "pdfplumber":pdf_pdfplumber}])], ignore_index=True)
    
```

```python
# 여기서 length 비교해서, 가장 길이가 긴 파일로 사용하면 될듯
df["length"] = df.apply(lambda x:[len(x["pypdf"]), len(x["pymupdf"]), len(x["pdfplumber"])], axis=1)
df["length"]
```

```python
from pprint import pprint as pp
col = df.iloc[0]
file_name = col["file_name"]
file_text = col["pypdf"]

with open(f"{file_name}.txt", "w", encoding='utf-8') as f:
    f.write(file_text)
```
