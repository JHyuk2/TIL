```python
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image_path = '/content/drive/MyDrive/camera/image_4.png'
# input_path = '/content/drive/MyDrive/train_sampled_500MB.csv'
```

```python
image_open = Image.open(image_path)
plt.imshow(image_open)
```

이미지 읽기

```python
# 이미지를 넘파이로 바꾸기 사이즈 보기
import numpy as np
np_image = np.array(image_open)

print(np_image.shape)
```

```python
# 이미지 사이즈부터 줄여보자
resized_image = image_open.resize(size=(256, 128))

# 사이즈를 만들 때 (W x H x 3) 사이즈로 구성되는 것 같다.
plt.imshow(resized_image)
```

```python
# 사진을 컬러 채널별로 만들어주기.
np_resized_image = np.array(resized_image)

resized_red = np_resized_image[:,:,0]
resized_green = np_resized_image[:,:,1]
resized_blue = np_resized_image[:,:,2]
```

```python
# 내가 원했던 건 grayscale인데, PIL에선 지원해주지 않는다...? (충격)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].imshow(np_resized_image, cmap='gray')
axes[0, 0].set_title('original')
axes[0, 0].axis('off')

axes[0, 1].imshow(resized_red, cmap='Reds')
axes[1, 0].imshow(resized_green, cmap='Greens')
axes[1, 1].imshow(resized_blue, cmap='Blues')
```
